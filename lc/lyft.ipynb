```python
class Job:
    def __init__(self, id, start_time, duration):
        self.id = id
        hour = int(start_time[:2])
        min = int(start_time[2:])
        self.start_time = hour * 60 + min
        self.end_time = self.start_time + int(duration)

path = './input.txt'
# input_file = open(path, 'r')
with open(path) as file:
    lines = [line for line in file]
n = int(lines[0])

##### workers ######
# index -> worker id
# val -> start/end time
worker_num = 0
worker_start_times = [0] * n
worker_end_times = [-1] * n
job_workers = {} # job id to worker id

##### jobs #####
jobs = {} # job id to job object
def construct_jobs():            
    id = 0
    for line in lines[1:]:
        start_time, duration = line.split(" ")
        new_job = Job(id, start_time, duration)
        jobs[id] = new_job
        id += 1


def job_schedule(worker_num):
    def next_available_worker(job_start_time, worker_end_times):
        for index, time in enumerate(worker_end_times):
            if job_start_time > time:
                return index


    sorted_jobs = [job for job in sorted(jobs.items(), key=lambda x: x[1].start_time)]
    for id, job in sorted_jobs:
        worker_id = next_available_worker(job.start_time, worker_end_times)
        worker_num = max(worker_num, worker_id)
        job_workers[id] = worker_id
        worker_start_times[worker_id] = job.start_time
        worker_end_times[worker_id] = job.end_time
    return worker_num

def construct_output(worker_num):    
    print(worker_num+1)
    sorted_output = [(job, worker) for job,worker in sorted(job_workers.items(), key=lambda x: x[0])]
    for job, worker in sorted_output:
        print("J" + str(job), "W" + str(worker))

##### main #####
construct_jobs()
worker_num = job_schedule(worker_num)
construct_output(worker_num)
```
