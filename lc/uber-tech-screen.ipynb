```python
# import requests
# import mysql.connector
# import pandas as pd

'''
input 1:
| O | E | E | E | X | 
| E | O | X | X | X |
| E | E | E | E | E |
| X | E | O | E | E |
| X | E | X | E | X | 

blocker_rows = [[4], [2,3,4], [], [0], [0,2,4]]

robot_col = 0
robot_col = 1
blocker_cols = [[3,4], [], [1,4], [1], [0,1,4]]

[['O','E','E','E','X'],['E','O','X','X','X'],['E','E','E','E','E'],['X','E','O','E','E'],['X','E','X','E','X']]

O = Robot, E = Empty, X = Blocker

input 2:
[2, 2, 4, 1] -> This means distance of 2 to the left closest blocker, 2 to the top closest blocker, 4 to the bottom closest blocker and 1 to the right closest blocker
'''

# n * m
# O(total cells) * O(2 * (n+m)) -> O(n*m) * O(n+m)

def find_robot(grid, distance):
    
    def get_distance(x, y):
        left, right = float('inf'), float('inf')
        for delta in range(1, m):
            if y - delta >= 0  and grid[x][y-delta] == 'X' or y - delta == -1:
                left = delta
                break
        for delta in range(1, m):
            if y + delta < m  and grid[x][y+delta] == 'X' or y + delta == m:
                right = delta
                break
        up, down = float('inf'), float('inf')
        for delta in range(1, n):
            if x - delta >= 0  and grid[x-delta][y] == 'X' or x - delta == -1:
                up = delta
                break
        for delta in range(1, n):
            if x + delta < n  and grid[x+delta][y] == 'X' or x + delta == n:
                down = delta
                break
        return [left, up, down, right]
            
    
    if not grid or not grid[0]:
        return None
    
    n, m = len(grid), len(grid[0])
    
    if n == m == 1:
        if grid[0][0] == 'O' and distance == [1, 1, 1, 1]:
            return (0, 0)
        else:
            return None
        
    
    for x in range(n):
        for y in range(m):
            if grid[x][y] == 'O':
                if get_distance(x, y) == distance:
                    return (x, y)
    
    return None
    
# input_00 = [['O','E','E','E','X'],['E','O','X','X','X'],['E','E','E','E','E'],['X','E','O','E','E'],['X','E','X','E','X']]
# input_10 = [2, 2, 4, 1]
# print(find_robot(input_00, input_10))

# input_01 = [['O','E','E','E','X'],['E','O','X','X','X'],['E','E','E','E','E'],['X','E','O','E','E'],['X','E','X','E','X']]
# input_11 = [1, 1, 3, 4]
# print(find_robot(input_01, input_11))

# input_02 = [['O']]
# input_12 = [1, 1, 1, 1]
# print(find_robot(input_02, input_12))

input_03 = [['X','E','E','E','X'],['E','X','X','X','X'],['E','E','E','E','E'],['X','E','X','E','E'],['X','E','X','E','X']]
input_13 = [1, 1, 1, 1]
print(find_robot(input_03, input_13))

```
