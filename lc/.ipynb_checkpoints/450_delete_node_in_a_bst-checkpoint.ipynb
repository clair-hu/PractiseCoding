```python
def insertToBST(root, n):
    if root:
        if root.value <= n.value:
            if root.right:
                insertToBST(root.right, n)
            else:
                root.right = n
        else:
            if root.left:
                insertToBST(root.left, n)
            else:
                root.left = n
```

```python
class Solution:
    def deleteNode(self, root: TreeNode, key: int) -> TreeNode:
        if not root:
            return None
        if key < root.val:
            root.left = self.deleteNode(root.left, key)
        elif key > root.val:
            root.right = self.deleteNode(root.right, key)
        else:
            if root.right == None:
                temp = root.left
                root = None
                return temp
            elif root.left == None:
                temp = root.right
                root = None
                return temp
            else:
                temp = root.right
                curr = temp
                while curr.left != None:
                    curr = curr.left
                curr.left = root.left
                return temp
        return root
```
