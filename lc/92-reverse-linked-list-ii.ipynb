```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    def printListNodes(self):
        print(self.val)
        if self.next:
            self.next.printListNodes()
def reverseBetween_1013(head, left, right):
    if left == right:
        return head
    dummy = ListNode(-1, head)
    i = 0
    curr = dummy
    while i < left:
        left_before = curr
        curr = curr.next
        i += 1
    while i < right:
        curr_after = curr.next
        curr_after_after = curr_after.next
        left_before_after = left_before.next
        left_before.next = curr_after
        curr_after.next = left_before_after
        curr.next = curr_after_after
        i += 1
    return dummy.next
    
```
