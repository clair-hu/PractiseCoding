```python
### bfs with iterative solution, passed
def largestValues(root):
    if not root:
        return []
    res = []
    nodes = [root]
    while nodes:
        cmax = -2**31
        next_nodes = []
        for n in nodes:
            cmax = max(cmax, n.val)
            if n.left:
                next_nodes.append(n.left)
            if n.right:
                next_nodes.append(n.right)
        res.append(cmax)
        nodes = next_nodes[:]
    return res


```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

node5 = TreeNode(5)
node3 = TreeNode(3)
node3p = TreeNode(3, node5, node3)
node9 = TreeNode(9)
node2 = TreeNode(2, None, node9)
node1 = TreeNode(1, node3p, node2)
largestValues(node1)
```
