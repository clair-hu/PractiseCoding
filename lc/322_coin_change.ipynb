```python
### iterative solution
def coinChange(coins, amount):
    dp = {}
    dp[0] = 0
    for target in range(1, amount+1):
        res = amount+1 # larger than result
        for coin in coins:
            if coin <= target and (target - coin) in dp:
                res = min(res, 1 + dp[target - coin])
        dp[target] = res
    return dp[amount] if dp[amount] <= amount else -1


import unittest
class TestCoinChange(unittest.TestCase):
    def test_0(self):
        coins = [1,2,5]
        amount = 11
        expected = 3
        self.assertEqual(coinChange(coins, amount), expected)

    def test_1(self):
        coins = [1,2,5]
        amount = 4
        expected = 2
        self.assertEqual(coinChange(coins, amount), expected)

    def test_2(self):
        coins = [5,7,9]
        amount = 4
        expected = -1
        self.assertEqual(coinChange(coins, amount), expected)

    def test_3(self):
        coins = [2,5,10,1]
        amount = 27
        expected = 4
        self.assertEqual(coinChange(coins, amount), expected)

if __name__ == '__main__':
    unittest.main()
```

```python
### rec -> memo solution - slower than iterative
def coinChange(coins, amount):

    ### memo initialization ###
    dp = {}
    dp[0] = 0
    ### memo initialization ###

    def rec(dp, target):
        if target == 0:
            return 0
        
        ### memo check existence ###
        if target in dp:
            return dp[target]
        ### memo check existence ###
            
        res = amount + 1 # largers than res
        for coin in coins:
            if coin <= target:
                res = min(res, rec(dp, target - coin)+1)
        
        ### memo add value ###
        dp[target] = res
        ### memo add value ###
        
        return res

    res = rec(dp, amount)
    return res if res <= amount else -1

import unittest
class TestCoinChange(unittest.TestCase):
    def test_0(self):
        coins = [2,4]
        amount = 3
        expected = -1
        self.assertEqual(coinChange(coins, amount), expected)

    def test_1(self):
        coins = [2,4]
        amount = 6
        expected = 2
        self.assertEqual(coinChange(coins, amount), expected)

    def test_3(self):
        coins = [1,2,5]
        amount = 11
        expected = 3
        self.assertEqual(coinChange(coins, amount), expected)

    def test_4(self):
        coins = [1,2,5]
        amount = 4
        expected = 2
        self.assertEqual(coinChange(coins, amount), expected)

    def test_5(self):
        coins = [5,7,9]
        amount = 4
        expected = -1
        self.assertEqual(coinChange(coins, amount), expected)

    def test_6(self):
        coins = [2,5,10,1]
        amount = 27
        expected = 4
        self.assertEqual(coinChange(coins, amount), expected)

if __name__ == "__main__":
    unittest.main()
```
