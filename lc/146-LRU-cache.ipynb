```python
### Passed!!!!

class Node:
    def __init__(self, key, val, prev = None, next = None):
        self.key = key
        self.val = val
        self.prev = prev
        self.next = next

class LRUCache:
    def __init__(self, capacity):
        self.capacity = capacity
        self.keyToNodes = dict()
        self.head, self.tail = None, None

    def put(self, key, value):
        if self.head == None or self.capacity == 1:
            self.keyToNodes = dict()
            dummyHead = Node(-1,-1, None, Node(-1, -1))
            dummyTail = Node(-1,-1, Node(-1, -1), None)
            node = Node(key, value, dummyHead, dummyTail)
            dummyHead.next = node
            dummyTail.prev = node
            self.keyToNodes[key] = node
            self.head, self.tail = node, node
            return
        
        if key in self.keyToNodes:
            self.moveToTail(key)
            self.keyToNodes[key].val = value
        else:
            if len(self.keyToNodes) == self.capacity:
                prevHead = self.head
                prevHeadNext = prevHead.next
                prevHeadNext.prev = Node(-1,-1, None, Node(-1, -1)) # dummy head
                self.head = prevHeadNext
                del self.keyToNodes[prevHead.key]
            prevTail = self.tail
            node = Node(key, value, prevTail, Node(-1,-1, Node(-1, -1), None))
            self.keyToNodes[key] = node
            self.tail = node
            prevTail.next = node


    def get(self, key):
        if key in self.keyToNodes:
            self.moveToTail(key)
            return self.keyToNodes[key].val
        return -1
    
    def moveToTail(self, key):
        node = self.keyToNodes[key]
        if self.tail == node:
            return
        prevNode = node.prev
        nextNode = node.next
        prevTail = self.tail
        self.tail = node
        node.prev = prevTail
        node.next = Node(-1,-1, Node(-1, -1), None) # dummy tail
        prevTail.next = node
        prevNode.next = nextNode
        nextNode.prev = prevNode
        if self.head == node:
            self.head = nextNode
    
    def printLRUCache(self):
        node = self.head
        while node:
            print(node.val, end=' ')
            node = node.next
        print()
```

```python
from collections import OrderedDict
class LRUCache:
    def __init__(self, capacity):
        self.orderedDict = OrderedDict()
        self.capacity = capacity
    
    def get(self, key):
        if key in self.orderedDict:
            self.orderedDict.move_to_end(key)
            return self.orderedDict[key]
        return -1
    
    def put(self, key, value):
        if key in self.orderedDict:
            self.orderedDict.move_to_end(key)
        elif len(self.orderedDict) == self.capacity:
            self.orderedDict.popitem(0)
        self.orderedDict[key] = value
        
```
