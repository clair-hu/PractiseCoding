```python
"""
You have a Database that stores unix timestamps as keys, and a list of filenames that were saved at that timestamp.
All of the filenames are unique.

You can interact with the Database *ONLY* using its two methods:

count_filenames(int start_time, int end_time)  // Counts filenames in timerange, inclusive of endpoints.
get_filenames(int timestamp) // Gets all filenames at a timestamp

The data from this Database was copied into another one, but one filename did not make it!
Your job is to find the missing filename. See test case at bottom of file.
"""
# Do NOT edit Database class code.
# Assume Database class methods are constant time when estimating time complexity.
class Database:
    def __init__(self) -> None:
        self.data = {}  # maps int to list of strings

    def get_filenames(self, timestamp):
        if timestamp in self.data:
            return self.data.get(timestamp)
        return []

    # Counts filenames in timerange, inclusive of endpoints.
    def count_filenames(self, start, end):
        count = 0
        for key in self.data.keys():
            if key >= start and key <= end:
                count += len(self.data.get(key))
        return count


# Returns the missing filename in the interval [start_time, end_time].
def find_missing_filename(db, db_copy, start_time, end_time):
    timestamps = []
    def get_timestamp_of_missing_file(start, end):
        mid = start + (end - start) // 2
        if start == end:
            timestamps.append(start)
            return

        if start < end: # todo check whether include equal
    
            db_cnt = db.count_filenames(start, mid)
            db_copy_cnt = db_copy.count_filenames(start, mid)
            if db_cnt != db_copy_cnt:
                get_timestamp_of_missing_file(start, mid)
            
            db_cnt = db.count_filenames(mid+1, end)
            db_copy_cnt = db_copy.count_filenames(mid+1, end)
            if db_cnt != db_copy_cnt:
                get_timestamp_of_missing_file(mid+1, end)
            
       
    get_timestamp_of_missing_file(start_time, end_time)
    files = []
    for timestamp in timestamps:
        db_files = db.get_filenames(timestamp)
        db_copy_files = db_copy.get_filenames(timestamp)
        for file in db_files:
            if file not in db_copy_files:
                files.append(file)

    return files


    
# left  = 2,  7,  7, 7
# right = 10, 10, 8, 7   
# mid   = 6,  8,  7, 7


def Test1() -> None:
    db = Database()
    db.data[2] = ["file1", "file2"]
    db.data[4] = ["file3"]
    db.data[5] = ["file4", "file5", "file6"]
    db.data[6] = ["file7", "file8", "file9", "file10"]
    db.data[7] = ["file11"]
    db.data[10] = ["file12"]

    db_copy = Database()
    #db_copy.data[2] = ["file1", "file2"]
    #db_copy.data[4] = ["file3"]
    #db_copy.data[5] = ["file4", "file5", "file6"]
    #db_copy.data[6] = ["file7", "file8", "file9", "file10"]
    #db_copy.data[7] = ["file11"]
    #db_copy.data[10] = ["file12"]

    # Should print [ "file2", "file11" ]
    print(find_missing_filename(db, db_copy, 2, 10))


Test1()

# n/2 n/2
# n/4 n/4 n/4 n/4
# ...
# 1 1 1
```
